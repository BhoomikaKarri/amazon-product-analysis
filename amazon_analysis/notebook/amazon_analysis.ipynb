```python
import pandas as pd

df = pd.read_csv("amazon.csv", encoding="latin-1")

df.head()
```

```python
df.head()
```

```python
import re

# clean discounted price
df["discounted_price"] = df["discounted_price"].apply(
    lambda x: float(re.sub(r"[^\d.]", "", str(x)))
)

# clean actual price
df["actual_price"] = df["actual_price"].apply(
    lambda x: float(re.sub(r"[^\d.]", "", str(x)))
)

# clean rating count
df["rating_count"] = df["rating_count"].astype(str).str.replace(",", "")
df["rating_count"] = df["rating_count"].astype(float)

df.head()
```

```python
df.dtypes
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")

df.dtypes
```

```python
# discount amount
df["discount_amount"] = df["actual_price"] - df["discounted_price"]

# price difference percentage check
df["price_difference_ratio"] = df["discount_amount"] / df["actual_price"]

# extract main category (before the first | )
df["main_category"] = df["category"].str.split("|").str[0]

df[["actual_price","discounted_price","discount_amount","main_category"]].head()

df["main_category"].value_counts()
```

```python
total_products = df.shape[0]

avg_rating = df["rating"].mean()

avg_discount = df["discount_percentage"].mean()

avg_price = df["discounted_price"].mean()

total_reviews = df["rating_count"].sum()

print("Total Products:", total_products)
print("Average Rating:", round(avg_rating,2))
print("Average Discount %:", round(avg_discount,2))
print("Average Product Price:", round(avg_price,2))
print("Total Reviews:", int(total_reviews))
```

```python
category_counts = df["main_category"].value_counts()

category_counts
```

```python
import matplotlib.pyplot as plt

category_counts.plot(kind="bar")

plt.title("Product Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.show()
```

```python
avg_rating_category = df.groupby("main_category")["rating"].mean().sort_values(ascending=False)

avg_rating_category
```

```python
avg_rating_category.plot(kind="bar")

plt.title("Average Product Rating by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")

plt.show()
```

```python
avg_discount_category = df.groupby("main_category")["discount_percentage"].mean().sort_values(ascending=False)

avg_discount_category
```

```python
avg_discount_category.plot(kind="bar")

plt.title("Average Discount Percentage by Category")
plt.xlabel("Category")
plt.ylabel("Average Discount (%)")

plt.show()
```

```python
plt.scatter(df["discount_percentage"], df["rating"])

plt.title("Discount vs Product Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Rating")

plt.show()
```
